Taking Iris data, plot the following with proper legend and axis labels: (Download IRIS data from:
https://archive.ics.uci.edu/ml/datasets/iris or import it from sklearn.datasets)

a. Plot bar chart to show the frequency of each class label in the data.

b. Draw a scatter plot for Petal width vs sepal width.

c. Plot density distribution for feature petal length.

d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

```python
import sklearn.datasets as skdata
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Plot bar chart to show the frequency of each class label in the data.

```python
iris = skdata.load_iris()
X = iris.data 
target = iris.target
names = iris.target_names

df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].replace(to_replace= [0, 1, 2], value = ['setosa', 'versicolor', 'virginica'])
```

```python
sns.barplot(df)
```

b. Draw a scatter plot for Petal width vs sepal width.

```python
sns.scatterplot(data=df, x='petal width (cm)', y='sepal width (cm)', hue='species')
```

c. Plot density distribution for feature petal length.

```python
df['petal length (cm)'].plot.density(color='green')
plt.title('Density plot for petal length (cm)')
plt.show()
```

d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

```python
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
```
