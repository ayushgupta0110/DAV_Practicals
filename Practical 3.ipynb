```python
import numpy as np
import pandas as pd
```

```python
# creating dataset
new_array = np.random.randn(50, 3)

n = int(np.count_nonzero(new_array) * .1)
 
# choosing random indexes to put NaN
index_nan = np.random.choice(new_array.size, n, replace=False)
 
# adding nan to the data.
new_array.ravel()[index_nan] = np.nan
print(new_array)
```

```python
data = pd.DataFrame(new_array, columns=['a', 'b', 'c'])
data
```

## a. Identify and count missing values in a dataframe.

```python
data.isnull().sum()
```

## b. Drop the column having more than 5 null values.

```python
col_name = data.isnull().sum()[data.isnull().sum() > 5].index
```

```python
data.drop(col_name, axis=1)
```

## c. Identify the row label having maximum of the sum of all values in a row and drop that row.

```python
max_index_row = data.sum(axis=1).idxmax()
print(max_index_row)
```

```python
data.drop(max_index_row, axis=0)
```

## d. Sort the dataframe on the basis of the first column.

```python
data.sort_values(by=data.columns[0])
```

## e. Remove all duplicates from the first column.

```python
data[data.columns[0]] = data[data.columns[0]].drop_duplicates()
```

```python
data
```

## f. Find the correlation between first and second column and covariance between second and third column.

```python
# correlation
data[data.columns[0]].corr(data[data.columns[1]])
```

```python
# covariance
data[data.columns[1]].cov(data[data.columns[2]])
```
