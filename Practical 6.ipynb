```python
import numpy as np
import pandas as pd
from random import sample
from dateutil.parser import parse
```

```python
dataset = pd.read_csv("weatherHistory.csv")
```

(A)  Compute mean of a series grouped by another series

```python
dataset[['Humidity']].groupby(dataset['Summary']).mean()
```

```python
dataset[['Temperature (C)']].groupby(dataset['Summary']).mean()
```

(B) Fill an intermittent time series to replace all missing dates with values of previous non-missing date.

```python
dataset['Formatted Date'].isnull().sum()
```

(C) Perform appropriate year-month string to dates conversion.

```python
temp = dataset.copy()
dummy_yearMonths = ['Apr 2020', 'Dec 2013', 'Mar 2017', 'Feb 2009', 'Jun 2001', 'June 2021']
temp_size = len(temp)
temp['Dummy_Date'] = [sample(dummy_yearMonths,1)[0] for i in range(temp_size)]
temp
```

```python
temp['Dummy_Date'].map(lambda d: parse(d))
```

(D) Split a dataset to group by two columns and then sort the aggregated results within the groups.

```python
def sort_values(dataset, column='Summary', ascending = True):
    return dataset.sort_values(by=column, ascending = ascending)
```

```python
temp = dataset.groupby(['Summary', 'Precip Type'])
temp1 = temp.apply(sort_values, column = 'Temperature (C)', ascending = False)
temp1
```

(E) Split a given dataframe into groups with bin counts.

```python
dataset.groupby(['Precip Type']).count()
```
